### Removing URL's
Write a function that removes URL's from a single tweet. 

**Function input:**
- A single string object (tweet) 

**Function output:**
- The tweet with URL's removed as a single object

````python
# define function that removes URL from single tweet
def remove_urls(text):
    pattern_url = r'http[s]?://(?:[A-Za-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9A-Fa-f][0-9A-Fa-f]))+\s'
    subs_url = r'url-web'
    text = train['message'].replace(to_replace = pattern_url, value = subs_url, regex = True)
    return text
````
